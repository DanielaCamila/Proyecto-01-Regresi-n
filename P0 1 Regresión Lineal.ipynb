### Marco Teórico

#### Regresión Lineal

- Es un método estadístico que modela la relación entre una variable dependiente (y) y una o más no dependientes (x).  
  Su forma más simple es con solo una variable dependiente y se representa de esta manera.  

  $$ y = \beta_0 + \beta_1x + \epsilon $$

- \( \beta_0 \) es el intercepto, \( \beta_1 \) es la pendiente y \( \epsilon \) es el término de error.  
  También existe la regresión múltiple, que tiene dos o más variables independientes, y se representa así:

  $$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n + \epsilon $$

- Se utiliza más que nada para predecir valores y analizar relaciones entre variables.  

#### Regresión Polinomial

- Es una extensión de la regresión lineal en la que la relación entre la variable independiente y la dependiente se modela como un polinomio de grado n.  
  Por ejemplo, una regresión polinomial de grado 2 se representa así:

  $$ y = \beta_0 + \beta_1x + \beta_2x^2 + \epsilon $$

- Este enfoque se usa cuando la relación entre variables no es lineal, sino curva o algún derivado del tipo.  
  Lo único a resaltar es que el modelo se puede **sobreajustar** si el grado del polinomio es demasiado alto.
  
#### Interacción de Factores  

- La interacción de factores ocurre cuando el efecto de una variable independiente sobre la variable dependiente depende del valor de otra variable independiente. Esto se representa de esta manera:  

  $$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \beta_3(x_1 \cdot x_2) + \epsilon $$  


- Aquí vemos cómo \( \beta_3 \) mide el efecto de la interacción entre 
\( x_1 \) y \( x_2 \).  


- Este concepto es importante en estudios donde se busca entender cómo dos variables influyen conjuntamente en el resultado.  

#### Significancia de Factores

- Es la importancia estadística de una variable independiente en el modelo de regresión. Se evalúa mediante pruebas de hipótesis, como el p_value, para determinar si un coeficiente es significativamente diferente de cero. Un p_value menor que un umbral (comúnmente 0.05) indica que la variable tiene un impacto significativo en la variable dependiente. 

#### Regularización (Ridge, Lasso, ElasticNet)
- **Ridge** añade una penalización L2 (suma de cuadrados de los coeficientes), por ende, no elimina variables solo reduce sus coeficientes evitando que dominen el modelo. Se recomienda utilizar cuando se sabe que todas las variables son relevantes y están correlacionadas, de esta manera tienes una pequeña contribución de todas las variables deseadas sin eliminar ninguna. 

$$ \text{RSS}_{L2} = \sum_{i=1}^n e_i^2  + \lambda \sum_{j=1}^p \hat{\beta_j}^2 $$

- **Lasso** añade una penalización L1 (suma de los valores absolutos de los coeficientes), por ende, puede hacer que algunos coeficientes sean exactamente cero, lo que equivale a seleccionar características importantes y descartar las irrelevantes. Es útil cuando se busca un modelo más sencillo y con menos variables. 

$$ \text{RSS}_{L1} = \sum_{i=1}^n e_i^2  + \lambda \sum_{j=1}^p |\hat{\beta_j}| $$

- **Elastic Net** combina las penalizaciones de L1 y L2, útil cuando hay muchas variables correlacionadas y necesitas una selección automática de características. 

 $$  
  J(\beta) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2  
  + \lambda_1 \sum_{j=1}^{p} |\beta_j|  
  + \lambda_2 \sum_{j=1}^{p} \beta_j^2  
  $$

```python
# Voy a quitar la columna de ciudades ya que no creo que aporte información importante al análisis, aparte de que se agregarían demasiadas dummies al modelo y quiero evitar eso
data = data.drop(columns = ["City", "id"])

# También quitar el texto de la columna de Sleep Duration y solo tomar el primer número que encuentre 
data["Sleep Duration"] = data["Sleep Duration"].str.extract(r'(\d+)').astype(float)
data = data.dropna()

# Separamos "x" y "y"
X = data.drop(columns = ["Work/Study Hours"])

y = pd.DataFrame()
y["Work/Study Hours"] = data["Work/Study Hours"]

numerical_features = ["Academic Pressure", "Work Pressure", "CGPA", "Sleep Duration", "Study Satisfaction", "Job Satisfaction", "Financial Stress", "Depression", "Age"]
categorical_features  = ["Dietary Habits", "Profession","Degree", "Have you ever had suicidal thoughts ?", "Family History of Mental Illness", "Gender"]

encoded_features = pd.get_dummies(X[categorical_features], 
               columns=categorical_features, 
               drop_first=False)

data_to_model_df = pd.DataFrame(X[numerical_features], 
                                columns=numerical_features).reset_index()

encoded_features_df = encoded_features.reset_index()

data_to_model = pd.concat([data_to_model_df, encoded_features_df], axis=1)

X = data_to_model.drop('index', axis=1)

# Terminamos de hacer las midificaciones necesarias a los datos, ahora dividimos en train y test

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.4, random_state = 2)

n = len(X_test)

# Una vez separados, escalamos los datos con la estandarización std

scaler = StandardScaler()
scaler.fit(X_train)

X_std_test = scaler.transform(X_test)
X_std_train = scaler.transform(X_train)
```

#### Con Penalización Ridge

```python
def regression_Ridge(X_std_train, X_std_test, y_train, y_test, alpha):

    from sklearn.linear_model import Ridge

    # Crear polinomios de grado "degree"
    X_train_ridge = X_std_train.copy()
    X_test_ridge = X_std_test.copy()

    # Modelo de Ridge
    ridge_model = Ridge(alpha= alpha)
    ridge_model.fit(X_train_ridge, y_train)

    # Predicciones
    y_pred_test_Ridge = ridge_model.predict(X_test_ridge)
    y_pred_test_Ridge = np.reshape(y_pred_test_Ridge, (-1, 1))
    
    y_pred_train_Ridge = ridge_model.predict(X_train_ridge)
    y_pred_train_Ridge = np.reshape(y_pred_train_Ridge, (-1, 1))

    # Parámetros del modelo
    intercepto_Ridge = ridge_model.intercept_
    betas_Ridge = ridge_model.coef_
    
    r2_test_Ridge = r2_score(y_test, y_pred_test_Ridge)
    
    r2_train_Ridge = r2_score(y_train, y_pred_train_Ridge)

    # Error estándar de los coeficientes
    n = len(X_train_ridge)
    p = X_train_ridge.shape[1]
    RSS_Ridge = np.sum((y_pred_test_Ridge - y_test)**2)
    sig_Ridge = RSS_Ridge / (n - p)

    X_inv = np.linalg.inv(X_test_ridge.T @ X_test_ridge + np.eye(p) * 0.001)
    std_error_Sp_Ridge = np.sqrt(np.diagonal(X_inv))

    # p-values
    ti = betas_Ridge / std_error_Sp_Ridge
    pvalue_bi = 2 * (1 - stats.t.cdf(np.abs(ti), n - p))

    # Resultados
    print("p_values = ", pvalue_bi)
    print("r2_test = ", r2_test_Ridge)
    print("r2_train = ", r2_train_Ridge)
    print('Intercepto = ', intercepto_Ridge)
    print('Coeficientes = ', betas_Ridge)
```

```python
regression_Ridge(X_std_train, X_std_test, y_train, y_test, 5)
```

#### Con Penalización ElasticNet

```python
def regression_ElasticNet(X_train, X_test, y_train, y_test, alpha, l1_ratio):
    
    from sklearn.linear_model import ElasticNet
    
    # Crear polinomios de grado "degree"
    X_train_EN = X_train.copy()
    X_test_EN = X_test.copy()

    # Modelo de ElasticNet
    model_EN = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)
    model_EN.fit(X_train_EN, y_train)

    # Predicción en test
    y_pred_test_EN = model_EN.predict(X_test_EN)
    y_pred_test_EN = np.reshape(y_pred_test_EN, (-1, 1))
    
    y_pred_train_EN = model_EN.predict(X_train_EN)
    y_pred_train_EN = np.reshape(y_pred_train_EN, (-1, 1))

    # Parámetros del modelo
    intercepto_EN = model_EN.intercept_
    betas_EN = model_EN.coef_
    
    r2_test_EN = r2_score(y_test, y_pred_test_EN)
    
    r2_train_EN = r2_score(y_train, y_pred_train_EN)

    # Error estándar de los coeficientes
    n = len(X_train_EN)
    p = X_train_EN.shape[1]
    RSS_EN = np.sum((y_pred_test_EN - y_test) ** 2)
    sig_EN = RSS_EN / (n - p)
    X_inv = np.linalg.inv(X_test_EN.T @ X_test_EN + np.eye(p) * 0.001)
    std_error_Sp_EN = np.sqrt(np.diagonal(X_inv))

    # t-estadísticos y p-values
    ti = betas_EN / std_error_Sp_EN
    pvalue_bi = 2 * (1 - stats.t.cdf(np.abs(ti), n - p))

    # Resultados
    print("p_values = ", pvalue_bi)
    print("r2_test = ", r2_test_EN)
    print("r2_train = ", r2_train_EN)
    print('Intercepto = ', intercepto_EN)
    print('Coeficientes = ', betas_EN)
```

### Regresión Polinomial 

#### Sin penalización

```python
def regression(X_std_train, X_std_test, y_train, y_test, degree):

    from sklearn.linear_model import LinearRegression

    # Crear polinomios de grado "degree"
    X_train = X_std_train.copy()
    X_test = X_std_test.copy()
    for i in range(2, degree + 1):
        X_train = np.hstack((X_train, X_std_train**i))
        X_test = np.hstack((X_test, X_std_test**i))

    # Modelo
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Predicciones
    y_pred_test = model.predict(X_test)
    y_pred_test = np.reshape(y_pred_test, (-1, 1))
    
    y_pred_train = model.predict(X_train)
    y_pred_train = np.reshape(y_pred_train, (-1, 1))

    # Parámetros del modelo
    intercepto = model.intercept_
    betas = model.coef_
    
    r2_test = r2_score(y_test, y_pred_test)
    
    r2_train = r2_score(y_train, y_pred_train)

    # Error estándar de los coeficientes
    n = len(X_train)
    p = X_train.shape[1]
    RSS = np.sum((y_pred_test - y_test)**2)
    sig = RSS / (n - p)

    X_inv = np.linalg.inv(X_test.T @ X_test + np.eye(p) * 0.001)
    std_error_Sp = np.sqrt(np.diagonal(X_inv))

    # p-values
    ti = betas / std_error_Sp
    pvalue_bi = 2 * (1 - stats.t.cdf(np.abs(ti), n - p))

    # Resultados
    print("p_values = ", pvalue_bi)
    print("r2_test = ", r2_test)
    print("r2_train = ", r2_train)
    print('Intercepto = ', intercepto)
    print('Coeficientes = ', betas)
```

```python
def regression_ElasticNet(X_train, X_test, y_train, y_test, degree, alpha, l1_ratio):
    
    from sklearn.linear_model import ElasticNet
    
    # Crear polinomios de grado "degree"
    X_train_EN = X_train.copy()
    X_test_EN = X_test.copy()
    for i in range(2, degree + 1):
        X_train_EN = np.hstack((X_train_EN, X_train**i))
        X_test_EN = np.hstack((X_test_EN, X_test**i))

    # Modelo de ElasticNet
    model_EN = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)
    model_EN.fit(X_train_EN, y_train)

    # Predicción en test
    y_pred_test_EN = model_EN.predict(X_test_EN)
    y_pred_test_EN = np.reshape(y_pred_test_EN, (-1, 1))
    
    y_pred_train_EN = model_EN.predict(X_train_EN)
    y_pred_train_EN = np.reshape(y_pred_train_EN, (-1, 1))

    # Parámetros del modelo
    intercepto_EN = model_EN.intercept_
    betas_EN = model_EN.coef_
    
    r2_test_EN = r2_score(y_test, y_pred_test_EN)
    
    r2_train_EN = r2_score(y_train, y_pred_train_EN)

    # Error estándar de los coeficientes
    n = len(X_train_EN)
    p = X_train_EN.shape[1]
    RSS_EN = np.sum((y_pred_test_EN - y_test) ** 2)
    sig_EN = RSS_EN / (n - p)
    X_inv = np.linalg.inv(X_test_EN.T @ X_test_EN + np.eye(p) * 0.001)
    std_error_Sp_EN = np.sqrt(np.diagonal(X_inv))

    # t-estadísticos y p-values
    ti = betas_EN / std_error_Sp_EN
    pvalue_bi = 2 * (1 - stats.t.cdf(np.abs(ti), n - p))

    # Resultados
    print("p_values = ", pvalue_bi)
    print("r2_test = ", r2_test_EN)
    print("r2_train = ", r2_train_EN)
    print('Intercepto = ', intercepto_EN)
    print('Coeficientes = ', betas_EN)
```

```python
regression_ElasticNet(X_train, X_test, y_train, y_test, 5, 5, .7)
```
